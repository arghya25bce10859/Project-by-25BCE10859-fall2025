```python
# Personal Expense Tracker
expenses = {}

def add_expense():
    try:
        category = input("Enter category (e.g., Food, Transport): ").strip().capitalize()
        amount = float(input("Enter amount spent in category: "))
        if category in expenses:
            expenses[category] += amount
        else:
            expenses[category] = amount
        print(f" {amount:.2f} added to '{category}'.")                                                
    except ValueError:
        print("Invalid input. Please enter a number to the expenses.")

def view_expenses():                                                  
    print("Expense Summary:")
    total = sum(expenses.values())
    if expenses:
        for category, amount in expenses.items():
            print(f" - {category}: {amount:.2f}")                                                        
        print("-" * 20)                                                                                  
        print(f" **TOTAL: {total:.2f}**")                                                                
    else:
        print("No expenses recorded yet.")

def main():                                                   
    print("Welcome to your Personal Expense Tracker!")
    while True:
        print("Choose an option:")
        print("1. Add an expense")
        print("2. View all expenses")
        print("3. Exit")
        option = input("Enter choice (1/2/3): ")
        if option == '1':
            add_expense()
        elif option == '2':
            view_expenses()
        elif option == '3':
            print("Thank you for tracking your expenses. Goodbye!")
            break
        else:
            print("Invalid choice. Please try again and choose between choices 1, 2, and 3.")

if __name__ == "__main__":
    main()
```
